### **`get_time_cap(board, time_caps)`**
- รับ `time_caps` เป็นลิสต์ของจำนวนเต็ม (มีขนาดอย่างน้อย 1 ช่องแน่นอน)
- รับ `board` เป็น numpy array (มีจำนวนแถวเป็นจำนวนเท่าของขนาดของ time_caps แน่นอน)
- คืน จำนวนเต็ม ที่หาได้ดังนี้
  -	แบ่งจำนวนแถวของ `board` ออกเป็นจำนวนช่วงเท่ากับขนาดของ `time_caps` 
  - หาหมายเลขแถวบนสุดของ `board` ที่มีบางช่องไม่เป็น 0
  - หมายเลขแถวน้อยสุดนี้ ตกอยู่ในเลขที่ช่วงเท่าใด ก็คืนจำนวนเต็มใน `time_caps` ที่เลขที่ช่วงนั้น
  - ในกรณี ที่เป็น `board` ว่าง ๆ  ให้คืนจำนวนเต็มช่องขวาสุดของ `time_caps`
- หมายเหตุ: ผลลัพธ์ของฟังก์ชันนี้แทน ช่วงเวลาที่รูปทรงอันใหม่จะอยู่ในตำแหน่งบนสุด ก่อนร่วงลงมา คือ ถ้ากระดานถูกเติมให้สูงมากขึ้น เวลาที่มีให้ตัดสินใจเพื่อเลื่อนวางในตำแหน่งที่เหมาะสมก็น้อยลง

### **`animate_drop(board, shape, c)`**

- รับ `board` และ `shape` เป็น numpy array 
- รับ `c` เป็นจำนวนเต็ม แทนเลขคอลัมน์ของ `board` โดย
 - `c` มีค่า 0 หมายถึงเลขคอลัมน์ซ้ายสุดของ `board`
 - `c` มีค่าได้ตั้งแต่ 0 จนถึงค่าผลต่างความกว้างของ `board` กับ `shape` <br>เช่น ถ้า `board` และ `shape` กว้าง 4 และ 3 ตามลำดับ `c` มีค่าได้แค่ 0 หรือ 1 แน่ ๆ
- คืนลิสต์ของบอร์ด โดยบอร์ดแรกสุดเป็นบอร์ดเมื่อ `shape` ถูกวางไว้ที่บนสุดของ `board` และช่องซ้ายสุดของ `shape` อยู่ที่คอลัมน์ `c` ของบอร์ด ส่วนบอร์ดถัดไปเป็นบอร์ด เมื่อ `shape` ขยับลงมาทีละแถวจนไม่สามารถลงต่อได้ <font color=red>แต่ถ้าไม่สามารถวาง `shape` ใน `board` ตอนเริ่มต้นได้ (เพราะมีอย่างอื่นขวางอยู่) ก็ต้องคืนลิสต์ว่าง</font>
- <font color=red>สำคัญมาก</font>: ห้ามเปลี่ยนแปลงข้อมูลใน `board` และ `shape` <br>(สามารถทำสำเนาด้วยคำสั่ง `new_board = np.array(board)` หรือ `new_shape = np.array(shape)`)

## ข้อห้าม (อ่านตรงนี้ให้เข้าใจอย่างละเอียด)
เพื่อให้สามารถประเมินความสามารถในการใช้งานบริการของ numpy จึงจำเป็นต้องกำหนดข้อห้ามดังนี้
- ห้ามเปลี่ยนบรรทัดแรกของ code cell
- ห้ามเปลี่ยนบรรทัด def ของฟังก์ชันที่ให้เขียน
- ฟังก์ชันต้องคืนข้อมูลและประเภทข้อมูลตามที่กำหนดเท่านั้น

- ฟังก์ชันต้องไม่เปลี่ยนแปลงข้อมูลภายในของพารามิเตอร์ทุกตัวที่ได้รับ
- ฟังก์ชันต้องไม่ print อะไรใด ๆ ทั้งสิ้น
- ห้าม import อะไรใด ๆ เพิ่ม (ยกเว้น numpy)
- <font color=red>ให้ใช้บริการของ numpy เฉพาะเท่าที่มีใน<a href="https://www.cp.eng.chula.ac.th/~somchai/2110101/2564-2/hw12/101-Python-v62-numpy.pdf">เอกสารประกอบการสอน</a>ของเทอมนี้เท่านั้น
- ห้ามใช้คำสั่ง `exec`, `eval` 
- ห้ามใช้คำสั่งจำพวก `for`, `while`, `map`, comprehension, recursive function
  - ในฟังก์ชัน
`get_score`, `get_time_cap`, `rotate_left`, `rotate_right`
  - สำหรับฟังก์ชัน `animate_drop` และ `animate_clear` อนุญาตให้ใช้คำสั่งข้างต้นอย่างใดอย่างหนึ่งได้อย่างมาก<u><b>หนึ่ง</b></u>ครั้ง</font>

````python
# HW12_NUMPY (ไม่ลบหรือแก้ไขบรรทัดนี้ หรือเพิ่มอะไรก่อนบรรทัดนี้ โดยเด็ดขาด)
# เขียนในเซลล์นี้เท่านั้น ถ้าต้องการเขียนฟังก์ชันเพิ่ม ก็เขียนในเซลล์นี้

import numpy as np

def get_score(board):
    return np.sum(np.sum(board!=0,axis=1)==board.shape[1])*40

def get_time_cap(board, time_caps):
    if np.sum(board)==0:
        return time_caps[-1]
    get_index = np.arange(0,board.shape[0])
    not_zero = np.sum(board,axis=1)!=0
    n = get_index[not_zero][0]//(board.shape[0]//len(time_caps))
    return time_caps[n]
    
def rotate_right(shape):
    return shape.T[:,::-1]

def rotate_left(shape):
    return shape.T[::-1,:]

def animate_drop(board, shape, c):
    # DON'T change any values in board and shape
    pos = []
    is_block = shape != 0
    r = 0
    count = 0
    # if np.sum(board[:shape.shape[0],c:shape.shape[1]+c]):
    #     return []
    while True:
        if r+shape.shape[0]>board.shape[0]:
            break
        copy = np.array(board)
        if count<shape.shape[0]:
            crop = copy[:count+1,c:c+shape.shape[1]]
            if np.sum(crop[is_block[shape.shape[0]-1-count:]])!=0:
                return []
            count+=1

        crop = copy[r:r+shape.shape[0],c:c+shape.shape[1]]

        if np.sum(crop[is_block])!=0:
            break
        else:
            crop = crop + shape
            copy[r:r+shape.shape[0],c:c+shape.shape[1]] = crop
            pos.append(np.array(copy))
            r+=1
    return pos

def animate_clear(board):
    # DON'T change any values in board
    pos = []
    is_not_full = (np.sum(board!=0,axis=1)==board.shape[1])==False
    cleared = is_not_full.reshape(board.shape[0],1)*np.array(board)
    pos.append(np.array(cleared))
    wat_row = np.arange(0,cleared.shape[0])
    while True:
        if np.sum(cleared)==0:
            break
        if len(wat_row[np.sum(cleared,axis=1)==0])==0:
            return []
        bottom_zero = wat_row[np.sum(cleared,axis=1)==0][-1]
        if len(wat_row[np.sum(cleared,axis=1)!=0])==0:
            return []
        above_nzero = wat_row[np.sum(cleared,axis=1)!=0][0]
        if bottom_zero<above_nzero:
            break
        cleared[above_nzero+1:bottom_zero+1,:] = cleared[above_nzero:bottom_zero,:]
        cleared[above_nzero,:] = np.zeros((1,cleared.shape[1]))
        pos.append(np.array(cleared))
        # print(above_nzero)
        # print(bottom_zero)
    return pos
````
