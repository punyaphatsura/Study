เราสามารถประมาณค่าของ sin(x) ได้โดยฟังก์ชัน 

$sin(x)=\sum_{k=0}^{m}\frac{(-1)^kx^{2k+1}}{(2k+1)!} \tag{1}$  

ซึ่งสามารถกระจายได้เป็น 

$sin(x)=\frac{(-1)^0x^{2\cdot0+1}}{(2\cdot0+1)!}+\frac{(-1)^1x^{2\cdot1+1}}{(2\cdot1+1)!}+\frac{(-1)^2x^{2\cdot2+1}}{(2\cdot2+1)!}+\dots +\frac{(-1)^mx^{2m+1}}{(2m+1)!}\tag{2}$

ตามรูปทั่วไปของสมการนี้จะใช้ค่า $m=∞$

เราจะสามารถประมาณค่า $sin(x)$ ได้ใกล้เคียงขึ้น เมื่อใช้จำนวนพจน์ในสมการที่ 2 มากขึ้น ยิ่งค่า $|x|$ มาก ยิ่งต้องใช้จำนวนพจน์มากขึ้น

ยกตัวอย่างเช่น หาก $|x|$ มีค่าสูงสุดเป็น 2 และปรับเปลี่ยนค่า $m$ เป็น $0,1,2$ จะได้กราฟด้านล่าง ซึ่งจะเห็นว่าความแตกต่างหรือความผิดพลาดในการประมาณ (ซึ่งหลังจากนี้จะเรียกว่า error) มีค่าน้อยมาก เส้น $m=2$ ทับกับเส้น $sin(x)$
