## Problem 1
Sketch the following signals
1.
$x(t) = \sin \left (\frac{\pi}{4}t + 20^\circ \right)$


2.
$x(t) = \begin{cases}t+2 , t \leq -2 \\ 0 , -2 \leq t \leq 2\\ t-2 , t \geq 2 \end{cases}$

3.
$x(t) = 2e^{-t}, 0 \leq t < 1$ and $ x(t+1) = x(t) $ for all $t$

4.
$x(t) = u(t) + 5u(t-1) -2u(t-2)$

5.
$x(t) = r(t) - r(t-1) - u(t-2)$




```
t = np.arange(-10,10,0.01)

def x2func(t):
    if t<=-2:
        return t+2
    elif t>=-2 and t<=2:
        return 0
    else:
        return t-2

def x3func(t):
    if 0 <= t < 1:
        return 2 * np.exp(-t)
    else:
        rec = t % 1
        return x3func(rec)


x1 = np.sin(np.pi * t / 4 + np.pi / 9)
x2 = [x2func(ra) for ra in t]
x3 = [x3func(t3) for t3 in t]
x4 = np.heaviside(t,1) + 5*np.heaviside(t-1,1) - 2*np.heaviside(t-2,1)
x5 = np.maximum(0, t) - np.maximum(0, t - 1) - np.heaviside(t - 2, 1)
plt.plot(t,x1)
plt.show()
plt.plot(t,x2)
plt.show()
plt.plot(t,x3)
plt.show()
plt.plot(t,x4)
plt.show()
plt.plot(t,x5)
plt.show()
```

For the discrete time signal x[n] shown in Figure below, sketch each of the following

1. $x[2-n] $
2. $x[3n-4] $
3. $x\left[ \frac{2}{3}n+1 \right]  $
4. $x\left[-\frac{n+8}{4} \right] $

8. $x[2-n] +x[3n-4] $

```
# x[n]
t = np.arange(-2,4,dtype='int')
x_t = np.array([-3,1,2,-2,3,-1])
plt.stem(t, x_t)
plt.title('x[n]')
plt.show()

def dtd(t,x_t,n):
    t_new = t[(t%1==0)].astype('int')
    x_new = x_t[(t%1==0)]
    plt.stem(t_new,x_new)
    plt.title(str(n))
    plt.show()


t1 = t*(-1) + 2
dtd(t1,x_t,1)

t2 = (t+4)/3
dtd(t2,x_t,2)

t3 = (t-1)*3/2
dtd(t3,x_t,3)

t4 = (-4) * t - 8
dtd(t4,x_t,4)

t5 = np.cbrt(t)
dtd(t5,x_t,5)

t6 = np.arange(-1,5)
x_6 = np.array([-1,3,-1,5,1,-3])
dtd(t6,x_6,6)
```

## Problem 9

Evaluate the following integrals

1. $\int_{-\infty}^{\infty} \left( \frac{2}{3}t-\frac{3}{2} \right) \delta(t-1) \,dt$

1. $\int_{-\infty}^{\infty} (t-1) \delta \left(\frac{2}{3}t- \frac{3}{2} \right) \,dt$

1. $\int_{-3}^{-2} \left[ e^{(-t+1)} + \sin \left( \frac{2\pi t}{3} \right) \right] \delta \left(t- \frac{3}{2} \right) \,dt$

1. $\int_{-3}^{2} \left[ e^{(-t+1)} + \sin \left( \frac{2\pi t}{3} \right) \right] \delta \left(t- \frac{3}{2} \right) \,dt$

# **Ans**

1. $f(t) = \frac{2}{3}t - \frac{3}{2}$ and $t_0 = 1$: $\int_{-\infty}^{\infty} \left( \frac{2}{3}t-\frac{3}{2} \right) \delta(t-1) \,dt = \frac{2}{3}(1)-\frac{3}{2} = -\frac{5}{6}$

1. $f(t) = t - 1$ and $t_0 = \frac{9}{4}$: $\int_{-\infty}^{\infty} (t-1) \delta \left(\frac{2}{3}t- \frac{3}{2} \right) \,dt = \frac{3}{2}(\frac{9}{4} - 1) = \frac{15}{8}$

1. $f(t) = e^{(-t+1)} + \sin \left( \frac{2\pi t}{3} \right)$ and $t_0 = \frac{3}{2} $ but $ \frac{3}{2} $ not in range $ (-2,-3)$: $\int_{-3}^{-2} \left[ e^{(-t+1)} + \sin \left( \frac{2\pi t}{3} \right) \right] \delta \left(t- \frac{3}{2} \right) \,dt = 0$

1. $f(t) = e^{(-t+1)} + \sin \left( \frac{2\pi t}{3} \right)$ and $t_0 = \frac{3}{2}$: $\int_{-3}^{2} \left[ e^{(-t+1)} + \sin \left( \frac{2\pi t}{3} \right) \right] \delta \left(t- \frac{3}{2} \right) \,dt = e^{(-3/2+1)} + \sin \left( \frac{2\pi }{3} * \frac{3}{2} \right) = e^{(-1/2)}$
