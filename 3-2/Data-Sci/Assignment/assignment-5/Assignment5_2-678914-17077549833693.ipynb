```python
df = pd.read_csv('https://github.com/pvateekul/2110446_DSDE_2023s2/raw/main/DataSet/hed2020_dataset.csv')
```

```python
df
```

```python
df = df[df.label == 'e']

df
```

```python
selected_variables = ['cap-color-rate', 'stalk-color-above-ring-rate']
df = df[selected_variables]
df
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')
df_impute = imputer.fit_transform(df)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_impute)

df_scaled
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5, random_state=2020)
kmeans.fit(df_scaled)
cluster_labels = kmeans.labels_

cluster_labels
```

```python
centroid = kmeans.cluster_centers_
centroid
```

```python
centroid_original_scale = scaler.inverse_transform(centroid)
centroid_original_scale
```
