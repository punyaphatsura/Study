```python
web_url = "https://en.wikipedia.org/wiki/Chulalongkorn_University"
```

```python
import requests
from bs4 import BeautifulSoup

wiki = requests.get(web_url)
soup_wiki = BeautifulSoup(wiki.text, 'html.parser')
```
