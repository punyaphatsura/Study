```python
import requests
from bs4 import BeautifulSoup

wiki = requests.get("https://en.wikipedia.org/wiki/Chulalongkorn_University")
soup_wiki = BeautifulSoup(wiki.text, 'html.parser')
```

```python
print(soup_wiki.prettify())
```

```python
print(len(soup_wiki.find_all(attrs={'class': "mw-cite-backlink"})))
```
