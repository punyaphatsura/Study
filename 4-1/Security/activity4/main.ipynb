```python
def print_table(dic: dict):
    num = len(dic)

    border = ''.join(['|-' for _ in range(num)]) + '|'
    header = ''
    value = ''

    sorted_value_dic = sorted(dic.items(), key=lambda x:x[1])[::-1]

    for c in sorted_value_dic:
        header += f'| {c[0]} '
        value += f'| {dic[c[0]]} '
    else:
        header += '|'
        value += '|'
        
    print(header)
    print(border)
    print(value)

print_table(dic)
```

```python
import nltk
from nltk.corpus import words
import itertools
import string

nltk.download('words')
word_list = set(words.words())

CIPHER_TEXT = "PRCSOFQX FP QDR AFOPQ CZSPR LA JFPALOQSKR. QDFP FP ZK LIU BROJZK MOLTROE"

clean_cipher_text = ''.join(CIPHER_TEXT.split('.'))

def check_word(word: str) -> bool:
    return word.lower() in word_list

def bruteforce():
    for way in itertools.permutations(string.ascii_uppercase):
        table = { string.ascii_uppercase[i] : way[i] for i in range(len(way))}
        
        decrypted_text = ''
        for c in clean_cipher_text:
            if c in table:
                decrypted_text += table[c]
            else:
                decrypted_text += c  # Keep non-alphabet characters as is
        
        is_all_correct = all([check_word(word) for word in decrypted_text.split()])
        
        if is_all_correct:
            print("Decrypted Text:", decrypted_text)
            print(table)
            break

bruteforce()
```

```python
import string

def vigenere_cipher(plaintext, key, mode='encrypt'):
    alphabet = string.ascii_uppercase
    key = key.upper().replace(" ", "")
    plaintext = plaintext.upper().replace(" ", "")
    
    key = (key * (len(plaintext) // len(key) + 1))[:len(plaintext)]
    
    cipher_text = []
    for p, k in zip(plaintext, key):
        if p in alphabet:
            p_idx = alphabet.index(p)
            k_idx = alphabet.index(k)
            if mode == 'encrypt':
                cipher_idx = (p_idx + k_idx) % 26
            elif mode == 'decrypt':
                cipher_idx = (p_idx - k_idx) % 26
            cipher_text.append(alphabet[cipher_idx])
        else:
            cipher_text.append(p)
    
    return ''.join(cipher_text)

plaintext = "HELLO WORLD"
key = "CAT"
cipher_text = vigenere_cipher(plaintext, key, mode='encrypt')
decrypted_text = vigenere_cipher(cipher_text, key, mode='decrypt')

print(f"Plaintext: {plaintext}")
print(f"Key: {key}")
print(f"Ciphertext: {cipher_text}")
print(f"Decrypted Text: {decrypted_text}")
```

```python
import matplotlib.pyplot as plt

# Data for each cryptographic method
block_sizes = [16, 64, 256, 1024, 8192, 16384]  # Block sizes in bytes

# Throughput data in 1000s of bytes per second processed (kB/s)
sha1_throughput = [328823.76, 942309.45, 1880853.46, 2272336.51, 2477855.56, 2447120.84]
rc4_throughput = [1052006.11, 1228872.18, 1283202.58, 1304914.23, 1295559.19, 1304463.29]
blowfish_throughput = [128945.07, 132636.27, 132508.62, 133295.15, 132205.00, 133607.21]

# DSA performance is different (operations per second for sign and verify at different key sizes)
dsa_sign_throughput = [28221.6, 14540.6, 5139.7]  # For 512 bits, 1024 bits, and 2048 bits
dsa_verify_throughput = [41622.5, 17975.1, 5526.3]  # For 512 bits, 1024 bits, and 2048 bits
dsa_bits = [512, 1024, 2048]  # Key sizes for DSA

# Plotting the data
plt.figure(figsize=(10, 6))

# SHA1, RC4, Blowfish
plt.plot(block_sizes, sha1_throughput, marker='o', label='SHA-1')
plt.plot(block_sizes, rc4_throughput, marker='o', label='RC4')
plt.plot(block_sizes, blowfish_throughput, marker='o', label='Blowfish CBC')

plt.xscale('log')
# plt.yscale('log')
plt.xlabel('Block Size (bytes)')
plt.ylabel('Throughput (kB/s)')
plt.title('Performance of Cryptographic Algorithms (SHA-1, RC4, Blowfish CBC)')
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

plt.show()


plt.figure(figsize=(10, 6))
# DSA Sign and Verify
plt.plot(dsa_bits, dsa_sign_throughput, marker='o', label='DSA Sign')
plt.plot(dsa_bits, dsa_verify_throughput, marker='o', label='DSA Verify')

plt.xscale('log')
plt.yscale('log')
plt.xlabel('Key Size (bits)')
plt.ylabel('Operations per Second')
plt.title('Performance of Digital Signature Algorithm (DSA)')
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

plt.tight_layout()
plt.show()
```
